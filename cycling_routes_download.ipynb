```python
import geopandas as gpd
import osmnx as ox
import matplotlib.pyplot as plt

# Definování oblasti zájmu (Plzeň)
place_name = "Plzen, Czech Republic"

# Stažení dat o všech cestách v dané oblasti
tags_all_paths = {'highway': ['path', 'footway', 'track']}
all_paths_gdf = ox.geometries_from_place(place_name, tags=tags_all_paths)

# Stažení dat o cyklistických trasách v dané oblasti
tags_cycling_paths = {'highway': ['cycleway', 'path'], 'cycleway': True}
cycling_paths_gdf = ox.geometries_from_place(place_name, tags=tags_cycling_paths)

# Zobrazení prvních několika řádků dat pro všechny cesty
print(all_paths_gdf.head())

# Zobrazení prvních několika řádků dat pro cyklistické cesty
print(cycling_paths_gdf.head())

# Kontrola, zda GeoDataFrame není prázdný
if not all_paths_gdf.empty:
    # Vykreslení mapy s cestami
    fig, ax = plt.subplots(figsize=(10, 10))
    all_paths_gdf.plot(ax=ax, color='gray', alpha=0.5, label='All Paths')
    
    if not cycling_paths_gdf.empty:
        cycling_paths_gdf.plot(ax=ax, color='blue', alpha=0.7, label='Cycling Paths')
    
    plt.title(f"Paths in {place_name}")
    plt.legend()
    plt.show()
else:
    print("No paths found.")
```
